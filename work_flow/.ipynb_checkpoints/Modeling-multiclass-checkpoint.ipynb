## Set up cross validation for model selection

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection
from sklearn.model_selection import GroupShuffleSplit, StratifiedKFold, cross_val_score, train_test_split, GridSearchCV
from sklearn import linear_model, svm, naive_bayes, neighbors, ensemble
from sklearn.metrics import precision_score, recall_score, precision_recall_curve,f1_score, fbeta_score, auc, plot_roc_curve
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

## Split data for training and testing

```python
# Function to split testing and training data
def get_field_split_data(df, test_size=.2, seed=42):
    '''This function splits the data by field ID to maintain spatial independence of train and test. 
    If we use points within the same field for training and testing we would basically 
    be testing our model with training points
    '''
    rs = np.random.RandomState(seed)
    
    total_fields = samples['field'].unique() 
    test_fields = rs.choice(total_fields, 
                           size=int(total_fields.shape[0] * test_size), 
                           replace=False)

    df_train = samples[~samples['field'].isin(test_fields)]
    df_test = samples[samples['field'].isin(test_fields)] 

    y_train, y_test = df_train['crop'], df_test['crop']
    X_train = df_train.drop(['field','geometry','crop'],axis=1) 
    X_test = df_test.drop(['field','geometry','crop'],axis=1)

    return X_train, X_test, y_train, y_test, df_train, df_test
```

```python
# Split data
X_train, X_test, y_train, y_test, field_train, field_test = get_field_split_data(samples)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

## Train model using two classifiers Knn and Random Forests

```python
#X_train, X_test, label_train, label_test = train_test_split(samples.iloc[:,1:7], samples['crop'], \
#                                                            test_size=0.3, stratify=y, random_state=41)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print("The score for kNN is")
print("Training: {:6.2f}%".format(100*knn.score(X_train, y_train)))
print("Test set: {:6.2f}%".format(100*knn.score(X_test, y_test)))
```

```python
## Print confusion matrix for kNN
crop_names = pd.read_csv('/Users/AuerPower/Metis/git/crop-classification/data/crop_id_list.csv')
knn_confusion = confusion_matrix(y_test, knn.predict(X_test))
#plt.figure(dpi=150)
sns.heatmap(knn_confusion, cmap=plt.cm.Blues, annot=True, square=True, fmt='g',
           xticklabels=crop_names['crop'],
           yticklabels=crop_names['crop'])

plt.xlabel('Predicted crop')
plt.ylabel('Actual crop')
plt.title('kNN confusion matrix');
```

```python
rf = ensemble.RandomForestClassifier(n_estimators=500)
rf.fit(X_train, y_train)
print("The score for rf is")
print("Training: {:6.2f}%".format(100*rf.score(X_train, y_train)))
print("Test set: {:6.2f}%".format(100*rf.score(X_test, y_test)))
```

```python
# # Print confusion matrix for random forests
rf_confusion = confusion_matrix(y_test, rf.predict(X_test))
plt.figure(dpi=100)
sns.heatmap(rf_confusion, cmap=plt.cm.Blues, annot=True, square=True, fmt='g',
            xticklabels=crop_names['crop'],
            yticklabels=crop_names['crop'])

plt.xlabel('Predicted crop')
plt.ylabel('Actual crop')
plt.title('Random Forest confusion matrix');

# plt.savefig("confusion_matrix_logit_iris")
```

## Model Selection Using Grid Search

```python
# Set random seed
np.random.seed(0)
```

```python
# Create a pipeline
pipe = Pipeline([("classifier", RandomForestClassifier())])

# Create dictionary with candidate learning algorithms and their hyperparameters
search_space = [{"classifier": [LogisticRegression()],
                 "classifier__penalty": ['l2','l1'],
                 "classifier__C": np.logspace(0, 4, 10)
                 },
                {"classifier": [LogisticRegression()],
                 "classifier__penalty": ['l2'],
                 "classifier__C": np.logspace(0, 4, 10),
                 "classifier__solver":['newton-cg','saga','sag','liblinear'] ##This solvers don't allow L1 penalty
                 },
                {"classifier": [RandomForestClassifier()],
                 "classifier__n_estimators": [100, 120, 300, 500, 800, 1200],
                 "classifier__max_depth":[5,8,15,25,30,None],
                 "classifier__min_samples_leaf":[1,2,5,10,15,100],
                 "classifier__max_leaf_nodes": [2, 5,10]
                },
               {"classifier": [KNeighborsClassifier()],
                 "classifier__n_neighbors": [2,4,8,16],
                 },
                {"classifier": [svm.SVC()],
                 "classifier__C": [0.001, 0.01, 10, 100, 1000],
                 "classifier__class_weight":['Balanced', None]
                 }]
```

```python
# Create grid search 
gridsearch = GridSearchCV(pipe, search_space, cv=5, verbose=0, scoring='f1_weighted', n_jobs=-1)

# Fit grid search
best_model = gridsearch.fit(X_train, y_train)
```

```python
# View best model results
print(best_model.best_estimator_.get_params()['classifier'])
print("The mean accuracy of the model is:", best_model.score(X_train, y_train))
```

```python
# Predict target with best model
best_model.predict(X_test)
```

## Loop through models and get F1 score

```python
lr_model = linear_model.LogisticRegression(class_weight="balanced", solver='lbfgs', multi_class='auto')
nb_model = naive_bayes.GaussianNB()
knn_model = neighbors.KNeighborsClassifier()
svc_model = svm.SVC(probability=True, gamma="scale")
rf_model = ensemble.RandomForestClassifier(n_estimators=500)
et_model = ensemble.ExtraTreesClassifier(n_estimators=100)
ada_model = ensemble.AdaBoostClassifier()

model_names = ["lr_model", "nb_model", "knn_model", "svc_model", "rf_model", "et_model", "ada_model"]
models = [lr_model, nb_model, knn_model, svc_model, rf_model, et_model, ada_model]
```

```python
## Loop through each type of model and record cv score
f1_scores = []
for m in models:
    model_score = cross_val_score(m, X_train, y_train, cv=3, scoring='f1_weighted')
    f1_scores.append(np.mean(model_score))

print(f1_scores)
```
